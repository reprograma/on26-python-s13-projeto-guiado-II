```python
import pandas as pd
```

```python
Spotify20002023 =  pd.read_csv("Best Songs on Spotify from 2000-2023.csv", delimiter= ";")#Quando o CSV for separado por ; usar sep = ";"
Spotify20002023 
```

```python
traducao_colunas = {
    'title': 'Title',
    'artist': 'Artist',
    'top genre': 'Top_Genre',
    'year': 'Year',
    'bpm': 'BPM',
    'energy': 'Energy',
    'danceability': 'Danceability',
    'dB': 'DB',
    'liveness': 'Liveness',
    'valance': 'Valence',
    'duration': 'Duration',
    'acousticness': 'Acousticness',
    'speechiness': 'Speechiness',
    'popularity': 'Popularity'
}
Spotify20002023.rename(columns=traducao_colunas, inplace=True)
Spotify20002023
```

```python
Spotify20002023.info()
```

```python
Spotify2000_2023 = "Best Songs on Spotify from 2000 2023.csv"
Spotify20002023.to_csv(Spotify2000_2023, index=False)
```
