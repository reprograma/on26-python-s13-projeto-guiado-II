```python
# Traduo das colunas

traducoes = {
    'Location': 'Localizao',
    'Number of Households (in 1000)': 'Nmero de famlias (1000)',
    'Percentage of households with pets': 'Porcentagem de famlias com Pets (%)',
    'Number of Pet Households (in 1000)': 'Nmero de famlias com Pets (1000)',
    'Percentage of Dog Owners': 'Porcentagem de donos de ces (%)',
    'Dog Owning Households (1000s)': 'Famlias com ces (milhares)',
    'Mean Number of Dogs per household': 'Mdia de ces por famlias',
    'Dog Population (in 1000)':  'Populao canina (1000)',
    'Percentage of Cat Owners': 'Porcentagem de donos de gatos (%)',
    'Cat Owning Households': 'Famlias com gatos',
    'Mean Number of Cats' : 'Mdia de gatos por famlias',
    'Cat Population': 'Populao felina'
}

# result = [(d['color'], d['value']) for d in dictionarylist]

novas_colunas = []
for chave, valor in traducoes.items():
    novas_colunas.append(valor)

print(novas_colunas)
```

```python
# atribuindo os novos nomes das colunas
df.columns = novas_colunas
df.head(10)
```

```python
#para ver as informaes do dataframe, quantidade de linhas e colunas
df.shape
```

```python

# Obter o valor mximo e mnimo do nmero de famlias com Pets:
valor_max = df['Nmero de famlias com Pets (1000)'].max()
valor_min = df['Nmero de famlias com Pets (1000)'].min()

print("Valor mximo do nmero de famlias com Pets (1000)")
print(valor_max)

print("Valor mnimo do nmero de famlias com Pets (1000)")
print(valor_min)


```

```python


df[['Nmero de famlias com Pets (1000)']].describe(percentiles=[0.99]).T

```

**TOP 5 CIDADES COM MAIS PETS**

```python
# 5 cidades que contm o maior nmero de famlias com pets

top5_cidades = df.loc[:, ["Localizao", "Nmero de famlias com Pets (1000)"]].sort_values(by='Nmero de famlias com Pets (1000)', ascending=False).head(5)
top5_cidades

```

```python
top5_cidades.plot(kind='bar',color = '#ad5bf4')


plt.xlabel('Cidades')
plt.ylabel('N de Famlias com pets')
plt.title('Top 5 Cidades / Nmero de famlias com pets')

plt.show()
```

**5 CIDADES AS FAMLIAS TM MAIS CES**

```python
# 5 cidades com mais famlias com gatos
top5_gatos = df.loc[:, ["Localizao", "Famlias com gatos"]].sort_values(by='Famlias com gatos', ascending=False).head(5)
top5_gatos
```

```python
# 5 cidados com mais famlias com ces
top5_caes = df.loc[:, ["Localizao", "Famlias com ces (milhares)"]].sort_values(by='Famlias com ces (milhares)', ascending=True).head(5)
top5_caes
```
