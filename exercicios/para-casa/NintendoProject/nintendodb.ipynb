```python
# Importar as bibliotecas necessrias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

- Leitura e tratamento do DF

```python
# Carregando  e lendo os dados do arquivo CSV
df = pd.read_csv('nintendofile.csv', encoding='latin-1')
df
```

- Leitura e tratamento do DF

```python
# Traduo dos nomes das colunas
df = df.rename(columns={
        'Position': 'Posio',
        'Game': 'Jogo',
        'Publisher': 'Editora',
        'Developer': 'Desenvolvedor',
        'VGChartz Score': 'Avaliao VGChartz',
        'Critic Score': 'Avaliao dos Crticos',
        'User Score': 'Avaliao dos Usurios',
        'Total Shipped': 'Total de Unidades Vendidas',
        'Release Date': 'Data de Lanamento',
        'Last Update': 'ltima Atualizao'
})
df
```

```python
# Excluso da coluna Avaliao VGChartz
df.drop(['Avaliao VGChartz'], axis=1, inplace=True)
```

```python
# Tratamento adicional dos dados (se necessrio)
# Converso de datas para formato datetime
df['Data de Lanamento'] = pd.to_datetime(df['Data de Lanamento'])
df['ltima Atualizao'] = pd.to_datetime(df['ltima Atualizao'])


df
```

- Filtragem de Linhas com Base em Condies

```python
# Filtrando os Games do Mrio Bros
df_filtrado = df[df['Jogo'].str.contains('Mario')]
df_filtrado
```

###  __Visualizaes / Insights__
    

- 	__Grfico de Barras para os Jogos mais Avaliados pelos Crticos:__

```python
# Classificar o DataFrame com base na avaliao dos crticos em ordem decrescente
jogos_mais_avaliados_criticos = df.sort_values(by='Avaliao dos Crticos', ascending=False)

# Classificar o DataFrame com base na avaliao dos usurios em ordem decrescente
jogos_mais_avaliados_usuarios = df.sort_values(by='Avaliao dos Usurios', ascending=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Selecionar os 10 jogos mais avaliados pelos crticos e classific-los do mais avaliado para o menos avaliado
top_10_criticos = jogos_mais_avaliados_criticos.head(10).sort_values(by='Avaliao dos Crticos', ascending=True)

# Escolher uma paleta de cores vibrantes
cores = sns.color_palette("hls", len(top_10_criticos))

# Criar um grfico de barras horizontal com o Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='Avaliao dos Crticos', y='Jogo', data=top_10_criticos, palette=cores)
plt.title('Top 10 Jogos mais Avaliados pelos Crticos')
plt.xlabel('Avaliao dos Crticos')
plt.ylabel('Jogo')
plt.show()
```

- Insigths: 

1. O grfico reflete diversidade nas avaliaes de jogos, destacando a subjetividade das crticas e a variedade de preferncias.

2. Destaque para "The Legend of Zelda: Breath of the Wild": O jogo lidera como o mais bem avaliado entre os 10 principais, indicando alta aclamao e referncia na indstria.





 - __Grfico de Disperso para os Jogos mais Avaliados pelos Usurios:__

```python
# Selecionar os 10 primeiros jogos mais bem avaliados pelos usurios
top_10_usuarios = jogos_mais_avaliados_usuarios.head(10)

# Personalizar uma paleta de cores Seaborn com cores vivas
cores = sns.color_palette("Set1", 10)

# Criar um grfico de disperso com o Seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(x=top_10_usuarios['Avaliao dos Usurios'], y=top_10_usuarios['Jogo'], hue=top_10_usuarios['Jogo'], palette=cores, s=100)
plt.title('Top 10 Jogos mais Bem Avaliados pelos Usurios', fontsize=14, weight='bold')
plt.xlabel('Avaliao dos Usurios')
plt.ylabel('Jogo')
plt.gca().invert_yaxis()  # Inverter a ordem para mostrar o jogo com a melhor avaliao no topo
plt.legend(title='Jogo', bbox_to_anchor=(1, 1))
plt.show()

```

- __Insigths:__

1. O grfico de disperso destaca os 10 jogos mais bem avaliados pelos usurios, revelando quais jogos so os mais populares entre o pblico.

2. A disperso de pontos indica que os jogos mais bem avaliados pelos usurios podem variar amplamente em termos de classificaes, sugerindo que as preferncias dos usurios so diversas e no necessariamente alinhadas.

 - __Grfico duplo de Pizza  e Coluna para 5 principais Editoras de Jogos do Nintendo Switch__

```python
# Agrupar os dados por editora e calcular a contagem de jogos por editora
jogos_por_editora = df_filtrado['Editora'].value_counts()

# Selecionar as 5 principais editoras com base na contagem de jogos
top_5_editoras = jogos_por_editora.head(5)

# Personalizar uma paleta de cores mais vibrantes
cores = sns.color_palette("tab10", len(top_5_editoras))

# Configurar o tamanho do grfico
plt.figure(figsize=(14, 6))

# Criar o grfico de pizza ( esquerda)
plt.subplot(1, 2, 1)
plt.pie(top_5_editoras, labels=top_5_editoras.index, autopct="%1.1f%%", colors=cores, startangle=90, shadow=True, textprops={'fontsize': 12})
plt.title('Distribuio de Jogos por Editora (Top 5)', fontsize=14, weight='bold')

# Criar o grfico de colunas ( direita)
plt.subplot(1, 2, 2)
sns.barplot(x=top_5_editoras.index, y=top_5_editoras, palette=cores)
plt.title('Distribuio de Jogos por Editora (Top 5)', fontsize=14, weight='bold')
plt.xlabel('Editora', fontsize=12)
plt.ylabel('Contagem de Jogos', fontsize=12)

# Ajustar o espaamento entre os grficos
plt.tight_layout()

# Exibir os grficos
plt.show()
```

- __Insigths:__ 

1. A Nintendo  a editora dominante na srie Mario Bros, com a maioria esmagadora dos jogos da srie sendo publicados por ela.
2. H um jogo com editora "Unknown", indicando que a editora deste jogo no est claramente identificada nos dados.Outras editoras, como Ubisoft e Sega, tambm tm participao na srie, embora em menor quantidade. Ubisoft tem dois jogos e a Sega tem um jogo na srie Mario Bros.

 - __Grfico Duplo  para jogos do Mrio Bros__

```python
# Filtrando os Games do Mrio Bros
df_filtrado = df[df['Jogo'].str.contains('Mario')]

# Ordenando o DataFrame pelos jogos mais vendidos
df_filtrado = df_filtrado.sort_values(by='Total de Unidades Vendidas', ascending=False)

# Selecionando os 10 jogos mais vendidos
top_10_vendidos = df_filtrado.head(10)

# Configurao do estilo do Seaborn
sns.set(style="whitegrid")

# Criao de subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Grfico de Pizza
df_filtrado['Editora'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, ax=axes[0])
axes[0].set_title('Distribuio de Editoras para Jogos do Mario Bros', fontsize=14, weight='bold')

# Grfico de Barras Horizontal
sns.barplot(x='Total de Unidades Vendidas', y='Jogo', data=top_10_vendidos, palette='viridis', ax=axes[1])
axes[1].set_xlabel('Total de Unidades Vendidas')
axes[1].set_ylabel('Jogo')
axes[1].set_title('Top 10 Jogos Mais Vendidos do Mario Bros', fontsize=14, weight='bold')

# Ajustar o espaamento entre os grficos
plt.tight_layout()

# Exibir os grficos
plt.show()


```

- __Insigths:__ 

1. A Nintendo  a editora dominante na srie Mario Bros, com a maioria esmagadora dos jogos da srie sendo publicados por ela.
2. "Mario Kart 8 Deluxe" lidera em termos de total de unidades vendidas, com mais de 45 milhes de unidades vendidas, o que  substancialmente maior do que os outros jogos apresentados.
