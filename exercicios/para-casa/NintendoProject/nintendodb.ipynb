Este projeto tem como objetivo analisar e tratar os dados relacionados aos jogos do Nintendo Switch, utilizando uma base de dados que foi baixada do Kaggle e traduzida para o português. A base de dados contém informações como a posição dos jogos nas vendas totais, o nome do jogo, o nome da editora, o desenvolvedor do jogo, a pontuação no VGChartz, a pontuação pelos críticos, a pontuação média do usuário, o número de unidades do jogo enviadas, a data de lançamento do jogo e a data da última atualização.

### ⚙️ __Processamento__


-  Função de processamento dos dados

```python
# Instalação biblioteca de visualização de gráficos seaborn
!pip install seaborn
```

```python
# Importar as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

- Leitura e tratamento do DF

```python
# Carregando  e lendo os dados do arquivo CSV
df = pd.read_csv('nintendofile.csv', encoding='latin-1')
df
```

- Leitura e tratamento do DF

```python
# Tradução dos nomes das colunas
df = df.rename(columns={
        'Position': 'Posição',
        'Game': 'Jogo',
        'Publisher': 'Editora',
        'Developer': 'Desenvolvedor',
        'VGChartz Score': 'Avaliação VGChartz',
        'Critic Score': 'Avaliação dos Críticos',
        'User Score': 'Avaliação dos Usuários',
        'Total Shipped': 'Total de Unidades Vendidas',
        'Release Date': 'Data de Lançamento',
        'Last Update': 'Última Atualização'
})
df
```

```python
# Exclusão da coluna Avaliação VGChartz
df.drop(['Avaliação VGChartz'], axis=1, inplace=True)
```

```python
# Tratamento adicional dos dados (se necessário)
# Conversão de datas para formato datetime
df['Data de Lançamento'] = pd.to_datetime(df['Data de Lançamento'])
df['Última Atualização'] = pd.to_datetime(df['Última Atualização'])


df
```

- Filtragem de Linhas com Base em Condições

```python
# Filtrando os Games do Mário Bros
df_filtrado = df[df['Jogo'].str.contains('Mario')]
df_filtrado
```

### 📈 __Visualizações / Insights__
    

- 	__Gráfico de Barras para os Jogos mais Avaliados pelos Críticos:__

```python
# Classificar o DataFrame com base na avaliação dos críticos em ordem decrescente
jogos_mais_avaliados_criticos = df.sort_values(by='Avaliação dos Críticos', ascending=False)

# Classificar o DataFrame com base na avaliação dos usuários em ordem decrescente
jogos_mais_avaliados_usuarios = df.sort_values(by='Avaliação dos Usuários', ascending=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Selecionar os 10 jogos mais avaliados pelos críticos e classificá-los do mais avaliado para o menos avaliado
top_10_criticos = jogos_mais_avaliados_criticos.head(10).sort_values(by='Avaliação dos Críticos', ascending=True)

# Escolher uma paleta de cores vibrantes
cores = sns.color_palette("hls", len(top_10_criticos))

# Criar um gráfico de barras horizontal com o Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='Avaliação dos Críticos', y='Jogo', data=top_10_criticos, palette=cores)
plt.title('Top 10 Jogos mais Avaliados pelos Críticos')
plt.xlabel('Avaliação dos Críticos')
plt.ylabel('Jogo')
plt.show()
```

- Insigths: 

1. O gráfico reflete diversidade nas avaliações de jogos, destacando a subjetividade das críticas e a variedade de preferências.

2. Destaque para "The Legend of Zelda: Breath of the Wild": O jogo lidera como o mais bem avaliado entre os 10 principais, indicando alta aclamação e referência na indústria.





 - __Gráfico de Dispersão para os Jogos mais Avaliados pelos Usuários:__

```python
# Selecionar os 10 primeiros jogos mais bem avaliados pelos usuários
top_10_usuarios = jogos_mais_avaliados_usuarios.head(10)

# Personalizar uma paleta de cores Seaborn com cores vivas
cores = sns.color_palette("Set1", 10)

# Criar um gráfico de dispersão com o Seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(x=top_10_usuarios['Avaliação dos Usuários'], y=top_10_usuarios['Jogo'], hue=top_10_usuarios['Jogo'], palette=cores, s=100)
plt.title('Top 10 Jogos mais Bem Avaliados pelos Usuários', fontsize=14, weight='bold')
plt.xlabel('Avaliação dos Usuários')
plt.ylabel('Jogo')
plt.gca().invert_yaxis()  # Inverter a ordem para mostrar o jogo com a melhor avaliação no topo
plt.legend(title='Jogo', bbox_to_anchor=(1, 1))
plt.show()

```

- __Insigths:__

1. O gráfico de dispersão destaca os 10 jogos mais bem avaliados pelos usuários, revelando quais jogos são os mais populares entre o público.

2. A dispersão de pontos indica que os jogos mais bem avaliados pelos usuários podem variar amplamente em termos de classificações, sugerindo que as preferências dos usuários são diversas e não necessariamente alinhadas.

 - __Gráfico duplo de Pizza  e Coluna para 5 principais Editoras de Jogos do Nintendo Switch__

```python
# Agrupar os dados por editora e calcular a contagem de jogos por editora
jogos_por_editora = df_filtrado['Editora'].value_counts()

# Selecionar as 5 principais editoras com base na contagem de jogos
top_5_editoras = jogos_por_editora.head(5)

# Personalizar uma paleta de cores mais vibrantes
cores = sns.color_palette("tab10", len(top_5_editoras))

# Configurar o tamanho do gráfico
plt.figure(figsize=(14, 6))

# Criar o gráfico de pizza (à esquerda)
plt.subplot(1, 2, 1)
plt.pie(top_5_editoras, labels=top_5_editoras.index, autopct="%1.1f%%", colors=cores, startangle=90, shadow=True, textprops={'fontsize': 12})
plt.title('Distribuição de Jogos por Editora (Top 5)', fontsize=14, weight='bold')

# Criar o gráfico de colunas (à direita)
plt.subplot(1, 2, 2)
sns.barplot(x=top_5_editoras.index, y=top_5_editoras, palette=cores)
plt.title('Distribuição de Jogos por Editora (Top 5)', fontsize=14, weight='bold')
plt.xlabel('Editora', fontsize=12)
plt.ylabel('Contagem de Jogos', fontsize=12)

# Ajustar o espaçamento entre os gráficos
plt.tight_layout()

# Exibir os gráficos
plt.show()
```

- __Insigths:__ 

1. A Nintendo é a editora dominante na série Mario Bros, com a maioria esmagadora dos jogos da série sendo publicados por ela.
2. Há um jogo com editora "Unknown", indicando que a editora deste jogo não está claramente identificada nos dados.Outras editoras, como Ubisoft e Sega, também têm participação na série, embora em menor quantidade. Ubisoft tem dois jogos e a Sega tem um jogo na série Mario Bros.

 - __Gráfico Duplo  para jogos do Mário Bros__

```python
# Filtrando os Games do Mário Bros
df_filtrado = df[df['Jogo'].str.contains('Mario')]

# Ordenando o DataFrame pelos jogos mais vendidos
df_filtrado = df_filtrado.sort_values(by='Total de Unidades Vendidas', ascending=False)

# Selecionando os 10 jogos mais vendidos
top_10_vendidos = df_filtrado.head(10)

# Configuração do estilo do Seaborn
sns.set(style="whitegrid")

# Criação de subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Gráfico de Pizza
df_filtrado['Editora'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, ax=axes[0])
axes[0].set_title('Distribuição de Editoras para Jogos do Mario Bros', fontsize=14, weight='bold')

# Gráfico de Barras Horizontal
sns.barplot(x='Total de Unidades Vendidas', y='Jogo', data=top_10_vendidos, palette='viridis', ax=axes[1])
axes[1].set_xlabel('Total de Unidades Vendidas')
axes[1].set_ylabel('Jogo')
axes[1].set_title('Top 10 Jogos Mais Vendidos do Mario Bros', fontsize=14, weight='bold')

# Ajustar o espaçamento entre os gráficos
plt.tight_layout()

# Exibir os gráficos
plt.show()


```

- __Insigths:__ 

1. A Nintendo é a editora dominante na série Mario Bros, com a maioria esmagadora dos jogos da série sendo publicados por ela.
2. "Mario Kart 8 Deluxe" lidera em termos de total de unidades vendidas, com mais de 45 milhões de unidades vendidas, o que é substancialmente maior do que os outros jogos apresentados.
