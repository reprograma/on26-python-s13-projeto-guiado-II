Tradução do cabeçalho e lado da mama

```python

df = pd.read_csv('breast-cancer.csv')

mapeamento = {'left': 'esquerda', 'right': 'direita'}


df['Breast'] = df['Breast'].replace(mapeamento)


df.to_csv('breast-cancer_traduzido.csv', index=False)

traducoes = {
    'ID': 'Id',
    'Class': 'Recorrência',
    'Age': 'Idade',
    'Menopause': 'Menopausa',
    'Tumor-size': 'Tamanho-Tumor',
    'Inv-nodes': 'Numero-Linfonodos',
    'Node-caps': 'Penetração-Capsula-Linfonodo',
    'Deg-malig': 'Grau-Malignidade',
    'Breast': 'Mama',
    'Breast-quad': 'Quadrante-Mama',
    'Irradiat': 'Radioterapia',
}
novas_colunas = []
for chave, valor in traducoes.items():
    novas_colunas.append(valor)

df.columns = novas_colunas
df
```

Início dos Gráficos

Gráfico 1 e 2

```python

# Ajuste do tamanho da figura 
plt.figure(figsize=(12, 5))  

# Dividir a figura em duas colunas
plt.subplot(1, 2, 1)  

cores = ['#FF69B4', '#9370DB', '#FFFF00', '#32CD32', '#FFA500']
contagem_idade = df['Idade'].value_counts().head()
plt.bar(contagem_idade.index, contagem_idade.values, color=cores)
plt.xlabel('Idade')
plt.ylabel('Quantidade')
plt.title('Incidência de Câncer de Mama por Idade')


# Gráfico de barras para a coluna 'Mama'
plt.subplot(1, 2, 2)  # Segunda coluna da figura

contagem_lado = df['Mama'].value_counts().head()
plt.bar(contagem_lado.index, contagem_lado.values, color=cores)
plt.xlabel('Mama')
plt.title('Existe um lado do seio que seja mais comum o surgimento do nódulo?')


plt.tight_layout()  


plt.show()
```

Neste primeiro gráfico podemos ver que as chances de surgimento de tumor maligno é maior após os 40 anos.
Também pode-se observar que após os 70 anos diminui drásticamente da incidencia do surgimento da doença.

No segundo gráfico eu quis verificar se tem um lado que seja mais comum o aparecimento do tumor, porém não há uma diferença considerável.



Gráfico 3

Este gráfico mapeia a localização do tumor na mama

```python
quadrante_counts = df['Quadrante-Mama'].value_counts()

# gráfico de pizza
plt.figure(figsize=(6, 6))
plt.pie(quadrante_counts, labels=quadrante_counts.index, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')

# Adicione um título ao gráfico
plt.title('Localização do Nódulo')

# Exiba o gráfico
plt.show()
```

Aqui é possível identificar que é mais comum que o nódulo esteja localizado no lado esquerdo do seio

Leitura do CSV

```python
import sqlite3
import csv
banco = sqlite3.connect("breast-cancer.db") 
cursor = banco.cursor()
```

```python
file = open("breast-cancer.csv", encoding= 'UTF-8') 

conteudo = csv.reader(file)
```

```python
cursor.execute ('\
    CREATE TABLE IF NOT EXISTS Mama (\
        Id INTEGER,\
        Recorrência TEXT,\
        Idade TEXT,\
        Menopausa TEXT,\
        Tamanho TEXT,\
        Linfonodos TEXT,\
        Capsula TEXT,\
        Malignidade INTEGER,\
        Mama TEXT,\
        Quadrante TEXT,\
        Radioterapia TEXT\
    )\
')
```

```python
inserir_conteudo = '''
    INSERT INTO Mama (
        Id, Recorrência, Idade, Menopausa,
        Tamanho, Linfonodos, Capsula,
        Malignidade, Mama, Quadrante, Radioterapia
    ) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
'''
```

```python

cursor.executemany(inserir_conteudo, conteudo)

cursor.executemany
selecionar_tudo = "SELECT * FROM Mama"
entradas = cursor.execute(selecionar_tudo).fetchall();

banco.commit();
banco.close();

file.close()
```
