```python
import csv
import sqlite3
```

```python
file = open("Salary_Data2.csv", encoding= 'UTF-8') 
```

```python
conteudo = csv.reader(file)

connection = sqlite3.connect("Dados_Salario.db")

cursor = connection.cursor()
```

```python
cursor.execute('''CREATE TABLE IF NOT EXISTS Dados_Salario (
     
    Age INT,
    Gender VARCHAR,
    Education_Level VAECHAR,
    Job_Title VARCHAR,
    Years_of_Experience INT,
    Salary FLOAT  
)''')
```

```python
inserir_conteudo = "INSERT INTO Dados_Salario(Age,Gender,Education_Level,Job_Title,Years_of_Experience,Salary)VALUES( ?, ?, ?, ?, ?, ? )"

cursor.executemany(inserir_conteudo, conteudo)
```

```python
selecionar_tudo = "SELECT * FROM Dados_Salario"
entradas = cursor.execute(selecionar_tudo).fetchall()
```

```python
for entrada in entradas:
  print(entrada)
```

```python
connection.commit()
connection.close()
```
