# Búfalos no Acre

Embora essa pecuária seja significativamente menor que a bovina, o manejo deste animal também necessita de desflorestamento para criação de pasto. Por isso vamos analisar sua correlação com o desmatamento no estado.

```python

import pandas as pd
import matplotlib.pyplot as plt

dados_bubalinos = {
    'Ano': [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022],
    'Bubalinos': [2457, 4141, 3313, 3257, 3343, 3377, 3461, 4209, 4680, 4219, 5188]
}

df_bubalinos = pd.DataFrame(dados_bubalinos)

dados_desmatamento = {
    'Ano': [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022],
    'Área desmatamento Acre': [305, 221, 309, 264, 372, 257, 444, 682, 706, 889, 847]
}

df_desmatamento = pd.DataFrame(dados_desmatamento)

correlation = df_bubalinos["Bubalinos"].corr(df_desmatamento["Área desmatamento Acre"])

plt.figure(figsize=(6, 4))
plt.scatter(df_bubalinos["Bubalinos"], df_desmatamento["Área desmatamento Acre"], label=f"Correlação: {correlation:.2f}")
plt.xlabel("Quantidade de Bubalinos")
plt.ylabel("Área de Desmatamento (Acre)")
plt.title("Correlação entre Pecuária de Bubalinos e Desmatamento no Acre (2012-2022)")
plt.legend()
plt.grid(True)
plt.show()

print(f"Correlação entre Bubalinos e Área de Desmatamento: {correlation:.2f}")
```

## Bovinos

Tem um destaque na pecuária do estado do Acre.

```python

import pandas as pd
import matplotlib.pyplot as plt

dados_bovinos = {
    'Ano': [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022],
    'Bovinos': [2634467, 2697489, 2799673, 2916207, 2998969, 2895028, 3303633, 3509682, 3802112, 4047283, 4635381]
}

df_bovinos = pd.DataFrame(dados_bovinos)

dados_desmatamento = {
    'Ano': [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022],
    'Área desmatamento Acre': [305, 221, 309, 264, 372, 257, 444, 682, 706, 889, 847]
}

df_desmatamento = pd.DataFrame(dados_desmatamento)

correlation = df_bovinos["Bovinos"].corr(df_desmatamento["Área desmatamento Acre"])

plt.figure(figsize=(6, 4))
plt.scatter(df_bovinos["Bovinos"], df_desmatamento["Área desmatamento Acre"], label=f"Correlação: {correlation:.2f}")
plt.xlabel("Quantidade de Bovinos")
plt.ylabel("Área de Desmatamento (Acre)")
plt.title("Correlação entre Pecuária de Bovinos e Desmatamento no Acre (2012-2022)")
plt.legend()
plt.grid(True)
plt.show()

print(f"Correlação entre Bovinos e Área de Desmatamento: {correlation:.2f}")
```

Uma correlação positiva é indicada por coeficientes de correlação próximos de +1. 
Isso significa que, à medida que uma variável aumenta, a outra tende a aumentar também, seguindo uma tendência linear positiva.

Uma correlação negativa é indicada por coeficientes de correlação próximos de -1. 
Isso significa que, à medida que uma variável aumenta, a outra tende a diminuir, seguindo uma tendência linear negativa.

Uma correlação neutra (ou ausência de correlação) é indicada por um coeficiente de correlação próximo de 0. 
Isso sugere que não há uma relação linear aparente entre as variáveis, e elas variam de forma independente uma da outra.

E a correlação perfeita positiva, seria representada por um coeficiente de +1.

# Conclusão

Com base na análise dos dados apresentados neste estudo, fica claro que existe uma correlação significativa entre a expansão da pecuária de bovinos e bubalinos e o aumento do desflorestamento no estado do Acre. Nossas descobertas revelam que à medida que a atividade pecuária se expande, ocorre um aumento notável na taxa de desmatamento na região. 
Essas descobertas podem ser úteis para orientar políticas de desenvolvimento agrícola e ambiental, com o objetivo de mitigar os impactos negativos da pecuária no ecossistema local.

No entanto, é importante ressaltar que correlação não implica causalidade, e mais pesquisas são necessárias para compreender completamente os fatores subjacentes a essa relação e para desenvolver estratégias eficazes de manejo sustentável.
