```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = 'spotify-2023.csv'
spotify_data = pd.read_csv(file_path, encoding='latin1')
```

```python
spotify_data.head(6)
```

```python
spotify_data.describe()
```

```python
# Verificar os nomes das colunas
print(spotify_data.columns)
```

```python
numeric_columns = ['streams', 'bpm', 'danceability_%', 'valence_%', 'energy_%', 'acousticness_%', 'instrumentalness_%', 'liveness_%', 'speechiness_%']
statistics = spotify_data[numeric_columns].describe()
print(statistics)
```

```python
plt.figure(figsize=(10, 6))
```

```python
plt.subplot(1, 2, 1)
plt.hist(spotify_data['streams'], bins=20, color='skyblue', edgecolor='black')
plt.title('DistribuiÃ§Ã£o de Streams')
plt.xlabel('Streams')
plt.ylabel('FrequÃªncia')
```

```python
plt.subplot(1, 2, 2)
plt.boxplot(spotify_data['energy_%'])
plt.title('Energia das MÃºsicas')
plt.ylabel('Energia (%)')
plt.tight_layout()
plt.show()
```

```python
songs_by_year = spotify_data['released_year'].value_counts().sort_index()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(songs_by_year.index, songs_by_year.values, marker='o', linestyle='-')
plt.title('DistribuiÃ§Ã£o de MÃºsicas por Ano de LanÃ§amento')
plt.xlabel('Ano de LanÃ§amento')
plt.ylabel('NÃºmero de MÃºsicas')
plt.grid(True)
plt.show()
```
