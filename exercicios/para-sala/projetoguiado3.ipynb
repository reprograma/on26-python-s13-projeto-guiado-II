```python
import pandas as pd
import matplotlib.pyplot as plt
```

Introdução

Escolhi este database pois foi o que me causou impacto. Devido a minha desinformação, eu não tinha o conhecimento de que além do controle territorial que acomete a população a constantes humilhações, também há uma destruição sistemática de estruturas Palestinas.

```python
df = pd.read_csv("demolitions_pse_isr_conflict_2004-01_to_2023-08.csv", sep="," , encoding= 'UTF-8') 
```

```python
df.shape
```

```python
df.info()
```

```python
soma_coluna = df['housing_units'].sum()

# Crie um gráfico de barras com a soma da coluna
plt.bar('housing_units', soma_coluna)
plt.title('Soma da Coluna "Unidades Habitacionais"')
plt.xlabel('Categoria')
plt.ylabel('Soma')
plt.show()
```

```python
soma_coluna1 = df['people_left_homeless'].sum()
soma_coluna2 = df['minors_left_homeless'].sum()

# Crie um gráfico de barras para as somas
categorias = ['people_left_homeless', 'minors_left_homeless']
somas = [soma_coluna1, soma_coluna2]

plt.bar(categorias, somas)
plt.title('Soma das Colunas Adultos/Idosos e Menores Desabrigados')
plt.xlabel('Categoria')
plt.ylabel('Soma')
plt.show()
```

```python
contagem_reason = df['reason_for_demolition'].value_counts()
razoes_mais_frequentes = 3
contagem_reason = contagem_reason.head(razoes_mais_frequentes)

plt.bar(contagem_reason.index, contagem_reason)
plt.title('"Razões" alegadas com frequência')
plt.xlabel('Razões')
plt.ylabel('Contagem')
plt.xticks(rotation=90)  # Rotação dos rótulos no eixo x para facilitar a leitura
plt.show()
```

Insights
De acordo com o gráfico, a principal "Razão" alegada é sobre o pretexto de construção ilegal, e o que ocorre é uma recusa sistemática das licenças de construções para palestinos na Cisjordânia ocupada.
Diante disso, ocorre demolições massivas e arbitrárias desde 2004 deixando inúmeras famílias vulneráveis e ameaçando a existência deste povo.
